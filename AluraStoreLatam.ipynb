### Importaci√≥n de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. An√°lisis de facturaci√≥n



```python
ingreso1 = tienda["Precio"].sum()
ingreso2 = tienda2["Precio"].sum()
ingreso3 = tienda3["Precio"].sum()
ingreso4 = tienda4["Precio"].sum()

print(f"Ingreso total Tienda : ${ingreso1:,.2f}")
print(f"Ingreso total Tienda2 : ${ingreso2:,.2f}")
print(f"Ingreso total Tienda3 : ${ingreso3:,.2f}")
print(f"Ingreso total Tienda4 : ${ingreso4:,.2f}")

```

# 2. Ventas por categor√≠a

```python
cat1 = tienda['Categor√≠a del Producto'].value_counts()
cat2 = tienda2['Categor√≠a del Producto'].value_counts()
cat3 = tienda3['Categor√≠a del Producto'].value_counts()
cat4 = tienda4['Categor√≠a del Producto'].value_counts()

# Mostrar resultados
print("üìä Categor√≠as m√°s vendidas - Tienda :\n", cat1, "\n")
print("üìä Categor√≠as m√°s vendidas - Tienda 2:\n", cat2, "\n")
print("üìä Categor√≠as m√°s vendidas - Tienda 3:\n", cat3, "\n")
print("üìä Categor√≠as m√°s vendidas - Tienda 4:\n", cat4, "\n")
```

# 3. Calificaci√≥n promedio de la tienda


```python
calificacion1 = tienda['Calificaci√≥n'].mean()
calificacion2 = tienda2['Calificaci√≥n'].mean()
calificacion3 = tienda3['Calificaci√≥n'].mean()
calificacion4 = tienda4['Calificaci√≥n'].mean()

print("tienda su calificacion es: ", calificacion1, "\n")
print("tienda 2 su calificacion es: ", calificacion2, "\n")
print("tienda 3 su calificacion es: ", calificacion3, "\n")
print("tienda 4 su calificacion es: ", calificacion4, "\n")
```

# 4. Productos m√°s y menos vendidos

```python
mas_vendidos1 = tienda["Producto"].value_counts()
mas_vendidos2 = tienda2["Producto"].value_counts()
mas_vendidos3 = tienda3["Producto"].value_counts()
mas_vendidos4 = tienda4["Producto"].value_counts()

# Mostrar top 5 m√°s vendidos y menos vendidos de cada tienda
print("Tienda - M√°s vendidos:\n", mas_vendidos1.head(), "\n")
print("Tienda - Menos vendidos:\n", mas_vendidos1.tail(), "\n")
print("Tienda 2 - M√°s vendidos:\n", mas_vendidos2.head(), "\n")
print("Tienda 2 - Menos vendidos:\n", mas_vendidos2.tail(), "\n")
print("Tienda 3 - M√°s vendidos:\n", mas_vendidos3.head(), "\n")
print("Tienda 3 - Menos vendidos:\n", mas_vendidos3.tail(), "\n")
print("Tienda 4 - M√°s vendidos:\n", mas_vendidos4.head(), "\n")
print("Tienda 4 - Menos vendidos:\n", mas_vendidos4.tail(), "\n")

```

# 5. Env√≠o promedio por tienda

```python
# Calcular promedio de costo de env√≠o por tienda
envio1 = tienda["Costo de env√≠o"].mean()
envio2 = tienda2["Costo de env√≠o"].mean()
envio3 = tienda3["Costo de env√≠o"].mean()
envio4 = tienda4["Costo de env√≠o"].mean()

# Mostrar resultados
print(f"Costo de env√≠o promedio - Tienda: ${envio1:.2f}")
print(f"Costo de env√≠o promedio - Tienda 2: ${envio2:.2f}")
print(f"Costo de env√≠o promedio - Tienda 3: ${envio3:.2f}")
print(f"Costo de env√≠o promedio - Tienda 4: ${envio4:.2f}")
```

# 6. GRAFICO

```python
import matplotlib.pyplot as plt
# Gr√°ficos para los productos m√°s vendidos (Top 5)
fig, axs = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle("üìà Top 5 productos m√°s vendidos por tienda", fontsize=16)

mas_vendidos1.head().plot(kind="bar", ax=axs[0, 0], title="Tienda 1", color="skyblue")
mas_vendidos2.head().plot(kind="bar", ax=axs[0, 1], title="Tienda 2", color="lightgreen")
mas_vendidos3.head().plot(kind="bar", ax=axs[1, 0], title="Tienda 3", color="salmon")
mas_vendidos4.head().plot(kind="bar", ax=axs[1, 1], title="Tienda 4", color="plum")

for ax in axs.flat:
    ax.set_ylabel("Cantidad Vendida")
    ax.set_xlabel("Producto")
    ax.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.subplots_adjust(top=0.9)
plt.show()
```

```python
import pandas as pd

url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)
ingresos_tiendas = {
    "Tienda 1": tienda1["Precio"].sum(),
    "Tienda 2": tienda2["Precio"].sum(),
    "Tienda 3": tienda3["Precio"].sum(),
    "Tienda 4": tienda4["Precio"].sum()
}
```

```python
import matplotlib.pyplot as plt

# Preparar datos para el gr√°fico
labels = list(ingresos_tiendas.keys())
sizes = list(ingresos_tiendas.values())
colors = plt.cm.Set2(range(len(labels)))

# Crear gr√°fico circular
plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors)
plt.axis('equal')  # Para que se vea circular
plt.title("üí∏ Participaci√≥n de ventas por tienda")
plt.show()
```

```python
import folium
from folium.plugins import HeatMap

ventas1 = tienda1["Precio"].sum()
ventas2 = tienda2["Precio"].sum()
ventas3 = tienda3["Precio"].sum()
ventas4 = tienda4["Precio"].sum()

print("Ingresos por tienda:")
print(f"Tienda 1: ${ventas1:,.2f}")
print(f"Tienda 2: ${ventas2:,.2f}")
print(f"Tienda 3: ${ventas3:,.2f}")
print(f"Tienda 4: ${ventas4:,.2f}")

m = folium.Map(location=[4.5709, -74.2973], zoom_start=6)

tiendas = [
    (tienda1, "Tienda 1", "red", ventas1),
    (tienda2, "Tienda 2", "blue", ventas2),
    (tienda3, "Tienda 3", "green", ventas3),
    (tienda4, "Tienda 4", "purple", ventas4)
]

for df, nombre, color, ingreso in tiendas:
    heat_data = [[row["lat"], row["lon"]] for index, row in df.iterrows()]
    HeatMap(heat_data, radius=10).add_to(m)

    lat_promedio = df["lat"].mean()
    lon_promedio = df["lon"].mean()

    folium.CircleMarker(
        location=[lat_promedio, lon_promedio],
        radius=8,
        color=color,
        fill=True,
        fill_opacity=0.9,
        popup=f"{nombre} - Ingreso: ${ingreso:,.2f}"
    ).add_to(m)

m

```
